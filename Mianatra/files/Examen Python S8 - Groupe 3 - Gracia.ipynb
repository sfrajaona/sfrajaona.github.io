# Pile ou face avec l'homme riche (Monte Carlo Method)
Un homme très riche vous fait jouer au jeu suivant. Vous lancer une pièce normale jusqu’à obtenir une face. Soit N le nombre de lancées nécéssaire. 
* Si vous obtenez face à la première lancée, vous gagnez 2\$.
* Si vous obtenez face à la deuxième, vous gagnez 4\$,...
* Si vous obtenez face à la n-ème, vous gagnez $2^n$ \$ .

Cette fois ci l'homme riche a fixé un plafond de $2^{20} \$$ (aux alentours de un million de dollars). Si vous arrivez a 20 lancees, l'homme riche vous paye $2^{20}\$$ et arrete le jeu. Combien de dollars osez vous payer pour jouer a ce jeu (c'est a dire calculer l'esperance de gain de ce jeu)?
